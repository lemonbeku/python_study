# **Constants**
Constants are fixed values that do not change during program execution.
### 1. Numeric Constants
Numbers used directly in a program<br/>
Examples
```python
123
98.6
```
Two Types:
- Integer -> whole numbers (123)
- Floating point -> decimal numbers (98.6)

### 2. String Constants
Text values and must be enclosed in quotes.<br/>
Example:
```python
"Hello World
```
Used when programs need to display messages to users.<br/>
Example:
```python
print("Hello World")
```

# **Reserved Words (Keywords)**
Reserved words are special words that Python uses for its own syntax. They already have a specific meaning in Python.<br/>
Example:
```python
class
del
else
lambda
break
in
from
```
You cannot used reserved words in variable names, function names, class names. <br/>
Example:
```python
class = 5
```
class is a python keyword that's why it cannot be used as a variable name

# **Variables**
A variable is a name used to store data in memory
```python
x = 12,2
```
What Python does:
Python will finds a free space in memory, labels it x then stores the value 12.2.

<hr/>

# **Assignment**
**Assignment Statements**
Assignment uses the equals sign (=).<br/>
Example:
```python
x = 12.2
```
`=` in programming DOESN'T mean equality like in math
It actually means "Take the value on the right and store it in the variable on the left".
```python
value -> variable
```

**Reassignment**<br/>
Variables can change values.
Example:
```python
x = 12.2
x = 100
```
First, x is going stores 12.2 as their values, then x is updated to 100. So, the old value is overwritten and x is now 100

# **Variable Naming Rules**
**Allowed**<br/>
- letters
- numbers
- underscores
Example like
```python
spam
eggs
spam23
_myvariable
```

**Not Allowed**<br/>
- Cannot start with a number
- Cannot use special characters
- Cannot use reserved words
Example like
```python
23spam
spam!
spam#
spam$
class
if
while
```

<hr/>

# **Case Sensitivity**
Python distinguishes uppercase and lowercase letters.<br/>
Example:
```python
spam
Spam
SPAM
```
There are three different variables. However, using similar names is bad practice because it confuses readers.<br/>
**Best practice is to use lowercase variable names**.

<hr/>

# **Using Variables in Expressions**
Variables can appear on the right side of an assignment.<br/>
Example:
```python
x = 2
x = x + 2
```
So, in this case Python retrieves value of x (2), then it adds 2, finaly it stores the result back to x. The result is x = 4
