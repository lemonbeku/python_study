## Variable Names

Variables store values in memory.

Example:

```python
x = 35
y = 12
z = x * y
print(z)
```

Python only cares that variable names are **consistent and unique**.

---

## Mnemonic Variable Names

**Mnemonic variables** are names that help humans understand the code.

Example:

Less clear:

```python
a = 35
b = 12
c = a * b
```

More clear:

```python
hours = 35
rate = 12
pay = hours * rate
```

Python treats both the same, but **humans prefer meaningful names**.

---

## Python Does Not Understand Meaning

Python **does not know what variable names mean**.

Example:

```python
pay = hours * hours
```

Python will run it even if the logic is wrong.

Mnemonic names only help **humans read and debug code**.

---

## Assignment Statement

`=` means **assign a value**, not mathematical equality.

Example:

```python
x = 5
```

Meaning: store **5** in `x`.

---

## Right Side Runs First

Python evaluates the **right side first**, then stores the result.

Example:

```python
x = x + 1
```

Steps:

1. Take current value of `x`
2. Add 1
3. Store the result back in `x`

This is called **incrementing**.
