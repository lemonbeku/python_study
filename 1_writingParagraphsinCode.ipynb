# **Basic Programming Patterns**
Most programs are built using three basic patterns.
## **1. Sequential Pattern**
The simplest pattern. Instructions run one after another in order.
Example:


````python
x = 2
print(x)
x = x + 2
print(x)
````

Execution order:
1. Assign 2 to x
2. Print x
3. Add 2 to x
4. Print the new value
Python executes code top to bottom.
Flow of Sequential Execution
Programs follow a simple flow:
```
Step 1
 ↓
Step 2
 ↓
Step 3
 ↓
Step 4
```
Older programming methods used flowcharts to visualize this sequence.

## **2. Conditional Pattern**
Conditional logic allows programs to make decisions.
The main keyword:
```python
if
```
Example:

````python
x = 5
if x < 10:
  print("Smaller")
print("Finish")
````

Execution process:
1. Assign 5 to x
2. Check condition x < 10
3. If true -> run the indented statement
4. Continue with the rest of the program
</br>

Rules:
- if statement asks a question
- The question evaluates to True or False
- The block after : runs only if the condition is true

### Indentation in Python
Python uses indentation to define blocks of code.
Example:

````python
if x < 10:
  print("smaller")
````

The indented line belongs to the if block.
If the condition is false, Python skips the block entirely.

### Multiple Conditional Checks
Example:

````python
x = 5
if x < 10:
  print("smaller")
if x > 20:
  print("bigger")
print("finish")
````

Evaluation:
- `x < 10` -> True -> executes
- `x > 20` -> False -> skipped

## **3. Repetition Pattern (Loops)**
Loops allow programs to repeat actions many times.
Two main loop keywords:
```
while
for
```
### While Loop
Example:

````python
n = 5

while n > 0:
    print(n)
    n = n - 1

print("Done")
````

Process:
1. Check condition `n > 0`
2. If true -> execute loop block
3. Update variable
4. Repeat until condition becomes false
### Iteration Variable
The variable controlling the loop is called the iteration variable
. Example:
```python
n = 5
```
Inside the loop:
```python
n = n - 1
```
Purpose:
- Ensures the loop eventually stops
### Infinite Loops
If the iteration variable is not updated, the loop may run forever.
Example:
```python
while n > 0:
  print(n)
```
Because n never changes. Programs should usually be designed so loops eventually terminate.
