## **What is Conditional Execution**
- Allows programs to make decisions
- Unlike sequential execution (runs everything in order)
- Chooses different paths: turn left / turn right / go straight
- Make programs appear intelligent

## **1. `if` Statement (One-Way Decision)**
The purpose is to run code only if condition is true
```python
if conditional:
  code
```
- Condition returns `True` or `False`
- If `True` -> runs the code
- If `False` -> skips


````python
x = 5
if x < 10:
  print("small")
````

## **2. `if-else` (Two-Way Decision)**
The purpose is to choose between two paths.
```python
if condition:
  code_if_true
else:
  code_if_false
```
- Only one block runs
- Never both


````python
x = 4

if x > 2:
  print("bigger")
else:
  print("smaller")
````

## **3. `if-elif-else` (Multi-Way Decision)**
The purpose is to handle multiple conditions
```python
if condition1:
  code1
elif condition2:
  code2
elif condition2:
  code3
else:
  default_code
```
- Checked from top to bottom
- Stops at first True
- Only 1 block executes

````python
x = 5

if x < 2:
  print("small")
elif x < 10:
  print("medium")
else:
  print("large")
````

- Order Matters

````python
x = 5
if x < 20:
  print("A")
elif x < 10:
  print("B")
````

- `else` is optional

````python
x = 12
if x < 2:
  print("small")
elif x < 10:
  print("medium")
#nothing will runs
````

- Only one runs. Even if multiple are true, only first is executed in one block, except if another block is added

## **4. Nested `if`**
The purpose is to make more detailed decisions
```python
if condition1:
  if condition2:
    code
```


````python
x = 42
if x > 1:
  if x < 100:
    print("valid range")
````

## **5. `try-except`**
The purpose is to prevent programs from crashing and handle errors safely
```python
try:
  risky_code
except:
  fallback_code
```
Try runs first:
- if success -> skip except
- if error -> jump to except

````python
s = "hello"
try:
  num = int(s)
except:
  num = -1

print(num)
````

````python
s = "123"
try:
  num = int(s)
except:
  num = -1

print(num)
````

- Prevents Traceback. No Crash. Program continues running
- Acts like Insurance. "Try this", but "if it fails -> do something else"
- Stops inside `try`

````python
try:
  print("A")
  error
  print("B") #never runs
except:
  print("Error")
````

- Best practice is to keep `try` small
- Only include risky lines

````python
inp = input("Enter number: ")
try:
  val = int(inp)
except:
  val = -1

if val > 0:
  print("Nice work")
else:
  print("Not a number")
````
